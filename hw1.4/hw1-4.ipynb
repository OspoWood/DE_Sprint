### 1. Сгенерировать выборку с распределением Рэлэя ( https://ru.wikipedia.org/wiki/Распределение_Рэлея ) используя numpy и scipy и построить на основе ее кривую плотности распределения на основе гистограммы из Matplotlib.

```python
import scipy.stats as sps
import numpy as np
import matplotlib.pyplot as plt
```

```python
ray = sps.rayleigh(loc=10, scale=5)
```

```python
collection_ray = ray.rvs(size=10000, random_state=123)
```

```python
loc, scale = sps.rayleigh.fit(collection_ray) 
```

```python
xl = np.linspace(collection_ray.min(), collection_ray.max(), 100)
```

```python
fig, axe = plt.subplots()
axe.hist(collection_ray, density=1, label="Sample")
axe.plot(xl, sps.rayleigh(scale=scale, loc=loc).pdf(xl), label="Adjusted Distribution")
axe.set_title("Distribution Fit")
axe.legend()
axe.grid()
```

### 2. Построение графика на по функции

#### Сгенерировать массивы по функции (x^2 + y^2 - 1)^3 - x^2 y^3 = 0 и построить на основе них график.



```python
import scipy.stats as sps
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
t = np.linspace(0,6.3,1000)
x = 16*(np.sin(t))**3
y = 13*np.cos(t)-5*np.cos(2*t)-2*np.cos(3*t)-np.cos(4*t)
plt.figure('SweetHeart')
plt.plot(x,y,'r')
plt.title('HEART',fontsize='large',color='red')
plt.show()
```

### 3. Анализ зависимостей

##### Найти зависимости представленные в датасете приложенном к заданию.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
age = []
sex=[]
bmi = []
children = []
smokers = []
region = []
charges=[]
with open("./dataset.txt", "r") as file:
    for row in file.readlines():
        split = row.split(",")
        age.append(int(split[0]))
        
        if (split[1]=='female'):
            sex.append(0)
        else:
            sex.append(1)
            
        bmi.append(float(split[2])) 
        
        children.append(int(split[3]))
        
        if (split[4]=='yes'):
            smokers.append(1)
        else:
            smokers.append(0)
            
        if (split[5] == 'northwest'):
            region.append(0)
        elif (split[5] == 'northeast'):
            region.append(1)
        elif (split[5] == 'southeast'):
            region.append(2)
        elif (split[5] == 'southeast'):
            region.append(3)
        else: region.append(4)   
            
        charges.append(float(split[6]))
        
```

```python
age = np.array(age)
sex=np.array(sex)
bmi = np.array(bmi)
children = np.array(children)
smokers = np.array(smokers)
region = np.array(region)
charges=np.array(charges)
```

```python
plt.scatter(age, charges)
plt.legend(['Корреляция'])
```

```python
plt.scatter(np.log(charges), bmi)
plt.grid() 
plt.legend(['Не зависимость'])
```

```python
plt.scatter(age, children)
plt.legend(['Типичные значения'])
```

```python
plt.scatter(charges[age<35], bmi[age<35], label = "<35")
plt.scatter(charges[age>=35], bmi[age>=35], label = ">=35")

plt.legend(['кластеры'])
plt.show()
```
